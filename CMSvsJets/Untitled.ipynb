```python
import matplotlib.pyplot as plt
import numpy as np


dados1 = np.random.randn(5000, 1)
bins = np.linspace(-5, 5, 10)


plt.hist(dados1, bins=bins)
plt.show()

```

```python
dados2 = np.random.randn(500, 1)

plt.hist(dados2, bins=bins)

plt.show()
```
